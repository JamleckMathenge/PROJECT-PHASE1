```python
Top_rated_directors = pd.read_sql("""SELECT person_id , movie_id, averagerating ,primary_name
                                     FROM movie_ratings
                                     JOIN directors
                                     USING (movie_id)
                                     JOIN persons
                                     USING (person_id)
                                     WHERE numvotes > 1000
                                     GROUP BY person_id
                                     ORDER BY averagerating DESC
                                     LIMIT 20;
                                     """,conn)
Top_rated_directors
```

```python
# Create the plot
fig, ax = plt.subplots(figsize=(10,8))

# Graph X vs Y
ax.bar(Top_rated_directors["primary_name"], Top_rated_directors["averagerating"])

# Set title
ax.set_title("TOP RATED DIRECTORS")

# Set labels for X and Y axes
ax.set_xlabel("DIRECTOR'S NAME")
ax.set_ylabel("MOVIE RATING")
ax.tick_params(axis='x', labelrotation=90)
# Set text of legend
ax.legend(["RATING"]);
```

IN THE BAR CHART ABOVE, WE CAN SEE THE BEST DIRECTORS(through each director's unique identifier-person_id).
